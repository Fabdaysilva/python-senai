```python
# Venda por Estado
graficos = sns.barplot(x='Estado',y= 'ValorVenda', data = df, orient='v', errorbar=None)

# O método sns.barplot vai retornar uma lista de sub
# métodos utilizando os containers para acessar o texto
# do rótulos para cada colun, desta forma o python entende 
# que ele deve adicionar um rótulo em cada coluna, de acordo com o 
# comando graficos.containers

graficos.bar_label(graficos.containers[0])
```

```python
# Venda por Vendedor

sns.barplot(x='Vendedor',y='ValorVenda', data = df, errorbar=None)
```

#### Exercicio 01
- Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.

```python
# importar a biblioteca
import matplotlib.pyplot as plt

# GErar os dados para inserir no grafico

x = range(1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y (Trê vezes X)')
print(y)


# Plotar o gráfico de linha
plt.plot(x, y)

# adicionar um título a um gráfico
plt.title('Draw a line', fontsize = 15, c= 'blue')

# Adicionar título do eixo y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionar título do eixo x
plt.xlabel('Valores de X', fontsize = 12)

# Altera as cores dos valores dos eixos
plt.xticks(fontsize = 8, c = 'green')
plt.yticks(fontsize = 8, c ='purple')
```

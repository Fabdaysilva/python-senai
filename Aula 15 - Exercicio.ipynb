#### Passo 04 - Dividir o dataset para criar as bases de treino
Colunas em duas variaveis (x,y)

```python
x = IceCream[['Temperature']] # duas chaves para criar uma lista
y = IceCream['Revenue']
```

#### Passo 05 - Importar a biblioteca Scikit Learn e dividir a base em treino e teste, 
sendo que o tamanho do teste será de 25%

```python
# Importar o comando rain_test_split para dividir o dataframe
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
                                                                 
```

```python
X_train.shape
```

```python
y_test.shape
```

### Importar o algoritimo LinearRegression da biblioteca Scikitlaern

```python
from sklearn.linear_model import LinearRegression
```

```python
# Este comando irá gerar a linha de tendência
regressor = LinearRegression(fit_intercept= True)
regressor.fit(X_train, y_train)
```

```python
# Este comando irá gerar a linha de tendência
# E calcular os coeficientes linear e angular

print('Coeficiente angular (m):', regressor.coef_)
print('Coeficiente linear (b):', regressor.intercept_)
```

#### Passo 06 - Executar o teste

```python
y_previsao = regressor.predict(X_test)

y_previsao
```

```python
y_test
```

#### Plotando os resultados

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.ylabel('Receita [Dólares]')
plt.xlabel('Temperatura [ºC]')
plt.title('Receita gerada vc. Temperatura (Vendas de sorvete)')
```

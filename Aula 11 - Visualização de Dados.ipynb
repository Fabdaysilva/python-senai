#### Importando as bibliotecas (Matplotlib)

```python
import matplotlib.pyplot as plt # permite a geração de gráficos

import numpy as np # permite trabalhar com matrizes e vetores
```

#### Primeiro Gráfico com Matplotlib

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Plotar o gráfico

plt.plot(x, y)
```

#### Modificando os tipo dos marcadores

```python
plt.plot(x, y, marker= '*')
```

#### Alterando as cores da linha

```python
plt.plot(x, y, marker= '*', c= 'b') # usar a primeira letra da cor da cor ou o nome da cor / pick picker para cores
```

```python
plt.plot(
    x, y,             # os dados do gráfico
    marker = '*',     # Tipo de marcador
    c = 'green',      # Cor da linha  
    ds = 'steps',     # A forma com que os pontos são conectados
    ls = '--',        # Estilo da linha 
    mec = 'blue'      # Cor dos marcadores          
) 
```

#### Matplolib - adicionando outros elementos
Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras
possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a
atenção do que realmente importa.

```python
# Adiciona dados ao gráfico
plt.plot(x, y)

# Adicionar titulo ao eixo x
plt.xlabel('Valores de X', fontsize = 12)

# Adicional titulo ao eixo y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionar titulo ao gráfico
plt.title('Titulo do gráfico', fontsize = 18, c= 'purple')

# Altera as cores dos valores dos eixos
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c ='orange')
```

#### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
z = np.cos(x) 
plt.plot (x,y) 
plt.plot (x,z)
```

```python
# O camando plt.subplot permite criar grupo de gráficos 
# plt.subplot (linhas, colunas, posição elemento)


plt.subplot(2,1,1) # Primeiro gráfico
plt.plot(x,y)  #Conjunto de dados do primeiro gráfico
plt.title('subplots')

plt.subplot(2,1,2) # Segundo gráfico
plt.plot(x,z) # Conjunto de dados do segundo gráfico
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')

plt.subplot(1,2,2)
plt.plot(x,z)
```

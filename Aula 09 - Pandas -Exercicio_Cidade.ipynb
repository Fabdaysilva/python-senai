```python
# Declarar as variaveis para acessar o arquivo

path = 'C:\\Users\\FIC\\Documents\\Fabiana Python\\003 - Arquivos\\'
file = 'dados-vendas.xlsx'

# Acessar os dados do arquivo e colocar em um dataframe

dados = pd.read_excel(path + file, sheet_name='Vendas')
```

```python
#Seleciona apenas a coluna Vendedor e remove as duplicadas
df_cidade = dados['Cidade'].drop_duplicates()

# Converter o Pandas Series em uma lista
lista_cidade = df_cidade.tolist()

lista_cidade
```

```python
# Estrutura para gerar os aruivos de forma automatica

for cidade in lista_cidade:
    df_final = dados[dados['Cidade']== cidade]
    path_New = path + 'relatorio-venda cidade\\'
    # Salvar o arquivo na pasta
    df_final.to_excel(path_New + cidade + '.xlsx',index = False)
```
